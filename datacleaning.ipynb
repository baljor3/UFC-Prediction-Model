```python
import time
import pandas as pd
import numpy as np
import re
```

```python
statsOffighters=pd.read_csv("stats.csv")
```

```python
columns = ["name","division","rank","str.ac","grap.ac","sig.str.land","sig.str.att","take.land","take.att",
           "sig.str.land.min","sig.str.abs.min","take.avg.15min","sub.avg.15min","sig.str.def",
          "take.def","knock.ratio","avg.time","sig.str.head","sig.str.head.percent","sig.str.body",
          "sig.str.body.percent","sig.str.leg","sig.str.leg.percent",
          "sig.str.position.stand","sig.str.position.clinch","sig.str.position.ground",
          "sig.str.position.stand.percent","sig.str.position.clinch.percent","sig.str.position.ground.percent",
           "win.ko","win.sub","win.dec","win.ko.percent","win.sub.percent","win.dec.percent",
          "wins","losses","age","leg.reach","reach","height"]
temp = pd.DataFrame(columns =columns,index =range(len(statsOffighters)))
statsOffighters=statsOffighters.sort_values(by=["division"])
statsOffighters=statsOffighters.reset_index()
statsOffighters.drop('index',inplace=True, axis=1)
```

```python
for i in statsOffighters[statsOffighters["division"].str.contains("[a-z]*champion")].index:
    k=statsOffighters.loc[statsOffighters.index[i],"division"] 
    k =re.sub("champion","",k)
    statsOffighters.loc[i,"division"] = k
```

```python
for i in statsOffighters[statsOffighters["division"].str.contains("[a-z]*division")].index:
    k=statsOffighters.loc[statsOffighters.index[i],"division"] 
    k =re.sub("division","",k)
    statsOffighters.loc[i,"division"] = k
```

```python
for i in range(len(statsOffighters)):
    for j in range(len(statsOffighters)):
        if statsOffighters.loc[i,"opponent"] == statsOffighters.loc[j,"name"] and statsOffighters.loc[i,"division"] == statsOffighters.loc[j,"division"]:
            temp.iloc[i,]=statsOffighters.iloc[j,0:41]
            break
```

```python
together=pd.concat([statsOffighters,temp],axis = 1)
together.to_csv("merge_data.csv")
```
